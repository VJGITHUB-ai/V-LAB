###    Dealing with missing data technique

 ### 1. Dropping missing values

```python
data1 = data.dropna()
```

```python
# finding the null values after the dropna() method
data1.isnull().sum()
```

```python
data.dropna(inplace=True, axis=0)
print(data.head())
print(data.isnull().sum())
```

### 2.Replacing missing values

```python
data = pd.read_csv(r"C:/users/varty/downloads/Vartya.csv", usecols =['Index','Customer Id','Company' , 'City' , 'Phone 1'])
data.head()
```

```python
#replacing the NAN value usinf mean()
mean_value = data['Index'].mean()
print(mean_value)
```

```python
data1 = data.fillna(mean_value)
```

```python
data1.head(15)
```

```python
#replacing the NAN value usinf median()
median_value =data['Index'].median()
data3 = data.fillna(median_value)
print(median_value)
```

```python
data3.head(5)
```

### Aggregating data with .groupby() in pandas
You can do that by combining the .groupby() method with a summary method of your choice.
when you going to apply groupby() you need to clean the data

```python
data = pd.read_csv(r"C:/users/varty/downloads/Vartya.csv", usecols =['Index','Customer Id','Company' , 'City' , 'Phone 1'])
data.head()
```

```python
mean_value = data['Index'].mean()
print(mean_value)
```

```python
data.fillna(mean_value).head()
```

```python
data.groupby('Index').mean().head()
```
