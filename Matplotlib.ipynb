### Matplotlib Area Plot

```python
import matplotlib.pyplot as plt

# Data
x = [1, 2, 3, 4, 5]
y1, y2 = [10, 20, 15, 25, 30], [5, 15, 10, 20, 25]

# Area Chart
plt.fill_between(x, y1, y2, color='BLUE', alpha=0.4, label='Area 1-2')
plt.plot(x, y1, label='Line 1', marker='o')
plt.plot(x, y2, label='Line 2', marker='o')

# Labels and Title
plt.xlabel('X-axis'), plt.ylabel('Y-axis'), plt.title('Area Chart Example')

# Legend and Display
plt.legend(), plt.show()
```

```python
import matplotlib.pyplot as plt

# Sample data
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
house_values = [200000, 210000, 215000, 220000, 225000, 230000, 235000, 240000, 245000, 250000, 255000, 260000]

# Plotting the area plot
plt.fill_between(months, house_values, color='YELLOW', alpha=0.7) 
plt.plot(months, house_values, color='green', alpha=0.9)  

# Adding titles and labels
plt.title('House values')
plt.xlabel('Month')
plt.ylabel('House Value (Rs)')
plt.xticks(rotation=30)  # Rotate x-axis labels for better readability

# Display the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Sample data: Replace this with real data from a source like an API or CSV
dates = pd.date_range(start='2024-08-01', periods=10, freq='D')  # Example dates
prices = np.random.uniform(low=0.5, high=1.5, size=len(dates))  # Example prices

# Plotting the area plot
plt.figure(figsize=(15, 6))
plt.fill_between(dates, prices, color='red', alpha=0.4)  # Fill area under the curve
plt.plot(dates, prices, color='Slateblue', alpha=0.9)  # Line on top of the filled area

# Adding titles and labels
plt.title('Worldcoin (WDC) Price Over Time')
plt.xlabel('Date')
plt.ylabel('Price (Rs)')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability

# Display the plot
plt.tight_layout()  # Adjust layout to fit labels and title
plt.show()
```

```python
import matplotlib.pyplot as plt

# Sample data
chocolates = ['Milk', 'Dark', 'White', 'Bittersweet']
cocoa_content = [30, 70, 0, 60]  
prices = [2.5, 3.0, 2.0, 3.5]   
ratings = [4.2, 4.7, 3.8, 4.5]    
calories = [200, 250, 180, 230]   

# Create a figure and axes
fig, axs = plt.subplots(2, 2, figsize=(10, 8))

# Plot 1: Cocoa Content
axs[0, 0].bar(chocolates, cocoa_content, color='brown')
axs[0, 0].set_title('Cocoa Content')
axs[0, 0].set_xlabel('Chocolate Type')
axs[0, 0].set_ylabel('Cocoa Content (%)')

# Plot 2: Price
axs[0, 1].bar(chocolates, prices, color='darkred')
axs[0, 1].set_title('Price')
axs[0, 1].set_xlabel('Chocolate Type')
axs[0, 1].set_ylabel('Price (USD)')

# Plot 3: Ratings
axs[1, 0].bar(chocolates, ratings, color='gold')
axs[1, 0].set_title('Ratings')
axs[1, 0].set_xlabel('Chocolate Type')
axs[1, 0].set_ylabel('Rating (Out of 5)')

# Plot 4: Calories
axs[1, 1].bar(chocolates, calories, color='orange')
axs[1, 1].set_title('Calories')
axs[1, 1].set_xlabel('Chocolate Type')
axs[1, 1].set_ylabel('Calories')

# Adjust layout
plt.tight_layout()

# Show plot
plt.show()

```
